```python
df = pd.read_csv('sdata/data.csv', encoding='gb18030')
```

```python
df.shape
```

```python
def make_label(df):
    df["symbol"] = df["1"].apply(lambda x: 1 if x>0 else 0)
```

```python
make_label(df)
```

```python
df
```

```python
X = df[['news']]
y = df.symbol
```

```python
X=X.astype(str)
```

```python
import jieba
```

```python
def wordscut(mytext):
    return " ".join(jieba.cut(mytext))
```

```python
X['cut_news'] = X.news.apply(wordscut)
```

```python
X.cut_news[:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
def get_custom_stopwords(stop_words_file):
    with open(stop_words_file,'r', encoding='UTF-8') as f:
        stopwords = f.read()
    stopwords_list = stopwords.split('\n')
    custom_stopwords_list = [i for i in stopwords_list]
    return custom_stopwords_list
```

```python
stop_words_file = "sdata/stop.txt"
stopwords = get_custom_stopwords(stop_words_file)
```

```python
stopwords[10:]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer(stop_words=frozenset(stopwords))#使用停用词
```

```python
term_matrix.shape
```

```python
max_df = 0.55 # 在超过这一比例的文档中出现的关键词（过于平凡），去除掉。
min_df = 1 # 在低于这一数量的文档中出现的关键词（过于独特），去除掉。
```

```python
vect = CountVectorizer(max_df = max_df,
                       min_df = min_df,
                       token_pattern=u'(?u)\\b[^\\d\\W]\\w+\\b',
                       stop_words=frozenset(stopwords))
```

```python
term_matrix = pd.DataFrame(vect.fit_transform(X_train.cut_news).toarray(), columns=vect.get_feature_names())#转化为向量
```

```python
from sklearn.naive_bayes import MultinomialNB#朴素贝叶斯
nb = MultinomialNB()
```

```python
from sklearn.pipeline import make_pipeline#使用pipe
pipe = make_pipeline(vect, nb)
```
